```python
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf
from sklearn import datasets
```

```python
x_data=datasets.load_iris().data
y_data=datasets.load_iris().target
x_data
```

```python
y_data
```

```python
np.random.seed(116)
np.random.shuffle(x_data)
np.random.seed(116)
np.random.shuffle(y_data)
tf.random.set_seed(116)
```

```python
#构建网络模型
#全连接层
model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(units=3,
                          activation="softmax",
                          kernel_regularizer=tf.keras.regularizers.l2())
])
```

```python
# 配置神经网络训练方法
#优化器使用Adam
model.compile(optimizer=tf.keras.optimizers.Adam(lr=0.1),
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=False),
              metrics=["sparse_categorical_accuracy"])
#metrics=["sparse_categorical_accuracy"]
#tf.keras.losses.SparseCategoricalCrossentropy(from_logits=False)
```

```python
# 模型训练
history=model.fit(x_data, y_data, batch_size=32, epochs=100,
          validation_split=0.2, validation_freq=1)
```

```python
model.summary()
```

```python
#设置黑体为默认字体
plt.rcParams["font.sans-serif"] = "SimHei"
#设置正常显示负号
plt.rcParams["axes.unicode_minus"] = False

loss = history.history['loss']
val_loss = history.history['val_loss']
plt.figure(figsize=(6,6))
plt.plot(loss, label='训练损失函数')
plt.plot(val_loss, label='测试损失函数')
plt.title('训练和测试损失函数图')
plt.legend()
plt.show()
```

```python
acc=history.history["sparse_categorical_accuracy"]
val_acc=history.history["val_sparse_categorical_accuracy"]
plt.figure(figsize=(6,6))
plt.plot(acc, label='训练准确率')
plt.plot(val_acc, label='测试准确率')
plt.title('训练和测试准确率图')
plt.legend()
plt.show()
```
